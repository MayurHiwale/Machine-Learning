### Univariate : one variable

```python
sns.stripplot(df)   #strip plot is used when we have data is in numerical format
```

```python
sns.stripplot(df['size'])
```

```python
sns.stripplot(y='tip',data=df)
```

```python
sns.stripplot(df['sex'])
```

### bivariate()

If we wants to find out relation between two variable then we are use bivariate analysis.

```python
sns.stripplot(x='day',y='tip',data=df)
```

```python
sns.stripplot(x='day',y='total_bill',data=df)
```

```python
sns.stripplot(x='sex',y='total_bill',data=df)
```

```python
sns.stripplot(x='day',y='tip',color=None, data=df)
plt.ylabel('tip')
plt.show()
```
